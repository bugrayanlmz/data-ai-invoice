```python
# src/document_ai.py
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/Users/bugrayanlmz/Desktop/data-ai-project/data-ai-invoice/credentials/data-ai-invoice-454117-86bae93a14c4.json"
from google.cloud import documentai_v1 as documentai

def process_document(file_path, project_id, location, processor_id):
    """
    Belgeyi Google Cloud Document AI ile işleyip dönen document nesnesini geri verir.
    """
    # Kimlik dosyasının yolu ayarlandı mı kontrol et (alternatif olarak Notebook'ta ayarlanabilir)
    if "GOOGLE_APPLICATION_CREDENTIALS" not in os.environ:
        raise EnvironmentError("Lütfen GOOGLE_APPLICATION_CREDENTIALS ortam değişkenini ayarlayın.")

    # Document AI client'ı oluştur
    client_options = {"api_endpoint": "eu-documentai.googleapis.com"}
    client = documentai.DocumentProcessorServiceClient(client_options=client_options)

    # İşlemci adını oluştur (örnek: projects/{project_id}/locations/{location}/processors/{processor_id})
    name = f"projects/{project_id}/locations/{location}/processors/{processor_id}"

    # Dosya içeriğini oku
    with open(file_path, "rb") as f:
        file_content = f.read()

    # Raw document nesnesi oluştur (PDF formatı için)
    raw_document = documentai.RawDocument(content=file_content, mime_type="application/pdf")

    # Process request oluşturma
    request = documentai.ProcessRequest(name=name, raw_document=raw_document)

    # İşlem yap ve sonucu al
    result = client.process_document(request=request)
    document = result.document
    return document

# Örnek kullanım:
if __name__ == "__main__":
    # Bu örnekte ortam değişkeni ile credentials ayarlandıktan sonra çalıştırabilirsin.
    project_id = "data-ai-invoice-454117"          # Google Cloud projenin ID'si
    location = "eu"      # Örneğin "us" veya "eu"
    processor_id = "1e0be339e088cbdc"        # Oluşturduğun Document AI işlemcisinin ID'si
    file_path = "/Users/bugrayanlmz/Desktop/data-ai-project/data-ai-invoice/notebooks/bugra_bilet.pdf"  # İşlenecek PDF dosyasının yolu

    doc = process_document(file_path, project_id, location, processor_id)
    # Örneğin, işlem sonucundaki entity’leri yazdır:
    print("Fatura İşleme Sonuçları:")
    for entity in doc.entities:
        print(f"Entity Türü    : {entity.type_}")
        print(f"Entity Metni   : {entity.mention_text}")
        print(f"Güven Skoru    : {entity.confidence}")
        print("-" * 40)
```
